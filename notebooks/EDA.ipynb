```python
# _,ax = plt.subplots(nrows=6,ncols=2,figsize=(15,50))
hash_map = {}
size = []
label =[]
for idx,val in enumerate(df['Category'].unique()):
    a=df[df['Category'].isin([val])]
    b=a['Sub-Category'].unique()
    b = b.tolist()
    label.append(b)
    c=df[df['Sub-Category'].isin(b)]
    d=c['Sub-Category'].value_counts()
    e=d.tolist()
    size.append( e)
    
```

```python
for i,j in zip(size,label):
    plt.pie(i, labels=j, autopct='%1.1f%%', shadow=True)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Display the pie chart
    plt.show()
```
